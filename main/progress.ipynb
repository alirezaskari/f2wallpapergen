```python
def create_wallpaper(quote, author, index):
    # Set wallpaper dimensions and properties
    width = 1072
    height = 1448
    dpi = 96
    bit_depth = 24

    # Create a black and white background
    background_color = (0, 0, 0)
    image = Image.new("RGB", (width, height), background_color)

    # Load the author's image (replace 'author_image_path' with the actual image file path)
    author_image_path = "author.png"  # Provide the path to the author's image
    if os.path.exists(author_image_path):
        author_image = Image.open(author_image_path)
        author_image = author_image.convert("L")  # Convert to grayscale
        author_image = author_image.resize((width, height // 2))
        image.paste(author_image, (0, 0))

    # Add the quote text
    draw = ImageDraw.Draw(image)
    font = ImageFont.truetype("arial.ttf", size=36)  # You can change the font and size
    text_color = (255, 255, 255)
    text_x = 20
    text_y = height // 2 + 20
    draw.text((text_x, text_y), f'"{quote}" - {author}', fill=text_color, font=font)

    # Save the wallpaper
    image.save(f"book_wallpaper_{index}.png", dpi=(dpi, dpi), bits=bit_depth)
```
